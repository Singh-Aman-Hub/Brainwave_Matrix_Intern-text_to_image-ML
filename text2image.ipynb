```python
from diffusers import StableDiffusionPipeline
import torch
import matplotlib.pyplot as plt

# Load the lightweight model
pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/sd-turbo", 
    torch_dtype=torch.float16,
    use_safetensors=True
)

# Move the model to CPU (for M1), or use MPS if available (Apple Silicon GPU)
pipe = pipe.to("mps")  # Use "cpu" if "mps" gives any error

# Function to generate and display image
def generate_image(prompt):
    image = pipe(prompt).images[0]
    plt.imshow(image)
    plt.axis("off")
    plt.title(prompt)
    plt.show()

# Try generating a test image
generate_image("india as a global power")
```
